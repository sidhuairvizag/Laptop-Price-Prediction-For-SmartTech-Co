```python
import pandas as pd
import pickle
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# Load Model
try:
    with open('pipe.pkl', 'rb') as file:
        model_pipeline = pickle.load(file)
    print("✅ Model Loaded Successfully!")
except:
    print("❌ pipe.pkl not found! Put it in the same folder.")

# Widgets
style = {'description_width': '150px'}
layout = widgets.Layout(width='400px')

company = widgets.Dropdown(options=['Apple','Dell','HP','Lenovo','Asus','Acer','MSI','Microsoft','Razer'], description='Company:', value='Apple', style=style, layout=layout)
typename = widgets.Dropdown(options=['Ultrabook','Notebook','Gaming','2 in 1 Convertible'], description='Type:', value='Ultrabook', style=style, layout=layout)
ram = widgets.IntSlider(value=8, min=4, max=64, step=4, description='RAM (GB):', style=style, layout=layout)
weight = widgets.FloatSlider(value=1.37, min=0.5, max=5.0, step=0.1, description='Weight (kg):', style=style, layout=layout)
touch = widgets.Dropdown(options=[0,1], description='Touchscreen:', value=0, style=style, layout=layout)
ips = widgets.Dropdown(options=[0,1], description='IPS Panel:', value=1, style=style, layout=layout)
ppi = widgets.FloatText(value=226.98, description='PPI:', style=style, layout=layout)
cpu = widgets.Dropdown(options=['Intel Core i5','Intel Core i7','Intel Core i3','Other Intel Processor','AMD Processor'], description='CPU:', value='Intel Core i5', style=style, layout=layout)
hdd = widgets.IntSlider(value=0, min=0, max=2000, step=128, description='HDD (GB):', style=style, layout=layout)
ssd = widgets.IntSlider(value=128, min=0, max=2000, step=128, description='SSD (GB):', style=style, layout=layout)
gpu = widgets.Dropdown(options=['Intel','Nvidia','AMD'], description='GPU:', value='Intel', style=style, layout=layout)
os = widgets.Dropdown(options=['Mac','Windows','Others/No OS/Linux'], description='OS:', value='Mac', style=style, layout=layout)

output = widgets.Output()

def predict(b):
    with output:
        clear_output()
        input_data = {
            'Company': company.value, 'TypeName': typename.value, 'Ram': ram.value,
            'Weight': weight.value, 'Touchscreen': touch.value, 'Ips': ips.value,
            'ppi': ppi.value, 'Cpu_brand': cpu.value, 'HDD': hdd.value,
            'SSD': ssd.value, 'Gpu_brand': gpu.value, 'os': os.value
        }
        df = pd.DataFrame([input_data])
        price = model_pipeline.predict(df)[0]
        print(f"🎉 Predicted Price: ₹{price:,.2f}")

button = widgets.Button(description="🔮 Predict Price", button_style='success', layout=widgets.Layout(width='300px', height='50px'))
button.on_click(predict)

# Display UI
display(HTML("<h2 style='color:blue;text-align:center'>🖥️ Laptop Price Predictor</h2>"))
display(widgets.HBox([widgets.VBox([company, typename, ram, weight, touch, ips]),
                      widgets.VBox([ppi, cpu, hdd, ssd, gpu, os])]))
display(button)
display(output)
```
